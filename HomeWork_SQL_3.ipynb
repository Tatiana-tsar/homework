```python
#--task1  (lesson4)
#-- Компьютерная фирма: Сделать view (название all_products_flag_300) для всех товаров 
#(pc, printer, laptop) с флагом, если стоимость больше > 300. 
#Во view три колонки: model, price, flag
request = """
create view all_products_flag_300 (model, price, flag) as 
( select model, price, 'PC' from pc where price>300
 union all
 select model, price, 'Printer' from Printer where price>300
 union all
 select model, price, 'Laptop' from Laptop where price>300
 )


"""
pd.read_sql_query(request, conn)
```

```python
#--task2  (lesson4)
#-- Компьютерная фирма: Сделать view (название all_products_flag_avg_price) для всех товаров 
#(pc, printer, laptop) с флагом, если стоимость больше cредней . 
#Во view три колонки: model, price, flag

request = """
create view all_products_flag_avg_price (model, price, flag) as 
(select model, price, 'PC' from pc where price> 
     (select avg(price) from
      (select price from pc   union select price from Printer  union select price from Laptop ) k1)
union all
   select model, price, 'Printer' from Printer pc where price> 
     (select avg(price) from
      (select price from pc   union select price from Printer  union select price from Laptop ) k2)
union all
   select model, price, 'Laptop' from Laptop pc where price> 
     (select avg(price) from
      (select price from pc   union select price from Printer  union select price from Laptop ) k3)
 )
 

"""
pd.read_sql_query(request, conn)
```

```python
#--task3  (lesson4)
#-- Компьютерная фирма: 
#    Вывести все принтеры производителя = 'A' со стоимостью выше средней по принтерам производителя = 'D' и 'C'. 
#    Вывести model
request = """
select pr.model, maker, price from Printer pr, Product p where p.model=pr.model and p.maker ='A' and price >  
(select avg(price) from
   (select price from printer pr, product p where p.model=pr.model and maker='D'
    union all
    select price from printer pr, product p where p.model=pr.model and maker='C'
    )p
  )

"""
pd.read_sql_query(request, conn)

```

```python
#-task4 (lesson4)
#-- Компьютерная фирма: Вывести все товары производителя = 'A' со стоимостью выше средней 
#по принтерам производителя = 'D' и 'C'. Вывести model

request = """
select model from 
(select p.model,price from Printer pr, Product p where p.model=pr.model and p.maker ='A' 
 union
 select p.model,price from Pc pr, Product p where p.model=pr.model and p.maker ='A' 
 union
 select p.model,price from Laptop pr, Product p where p.model=pr.model and p.maker ='A' 
) f
where 
  price >  
   (select avg(price) from
   (select price from printer pr, product p where p.model=pr.model and (maker='D' OR maker='C')
    )p
  )
group by model

"""
pd.read_sql_query(request, conn)

```
