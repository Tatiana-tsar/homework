```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
import numpy as np
from scipy import stats
from tqdm.auto import tqdm


import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv('test_cluster.csv', encoding='cp1251', sep=';')
df.head()
```

```python
df.info()
```

```python
#min_casa_balance_1q показатели по счетам за квартал
#для людей возраста меньше 50 лет
#
#
sns.set(style='darkgrid')  
sns.relplot(x='age', y='min_casa_balance_1q', data=df.query("10<age<50"), kind='line')
plt.title('Линейная диаграмма', size=20, color='g'); # заголовок
plt.xlabel('Возраст')
plt.ylabel('показатели по счетам за квартал');
#plt.legend(labels=['x2','x3'])


```

```python
#показатели по счетам за квартал
#для людей возраста меньше 50 лет с разбивкой по полам
#
#
sns.relplot(x='age', y='min_casa_balance_1q', hue='gender', data=df.query("age<70"), kind='line')
plt.title('Линейная диаграмма с параметром hue=пол', size=20, color='g');
#
```

```python
# Длительность взаимоотношений с банком людей до 50 лет, с разбивкой по типам городов проживания клиентов
#
sns.relplot(x='age', y='full_mob', hue='city_type', data=df.query("10<age<50"), kind='line')
plt.title('Линейная диаграмма c параметром hue=city_type (типы городов)', size=20, color='g')
plt.xlabel('Возраст')
plt.ylabel('Длительность взаимоотношений с банком');
```

```python
sns.relplot(x='age', y='full_mob', hue='city_type',
            data=df.query("10<age<50"), kind='line', err_style="bars", ci=68)
plt.title('Линейная диаграмма с разбивкой по городам  стиль, bar', size=20, color='g')

```

```python
#создадим таблицу 
#age возраст
#gender пол repexc_date
#ml_balance ипотека
#cl_balance_0m потребительский кредит
#loan_balance_0m все кредиты
#td_balance_0m депозит
#casa_balance_0m счета
#nbi доход от клиента без OPEX

long_df = pd.melt(df, var_name='Item', value_name='Sales',
            id_vars=['age', 'gender','ml_balance', 'cl_balance_0m', 'td_balance_0m', 'casa_balance_0m', 'nbi'])
long_df.shape         
long_df.sample(n=5) 

```

```python
long_df.info
```

```python
# разбивка по возрасту, в зависимости от дохода от клиента без OPEX

g = sns.relplot(x='age', y='nbi', data=long_df, kind='line', style="event")
```

```python
# разбивка на графики по полу
g = sns.relplot(x='age', y='nbi', data=long_df,col="gender")
```

```python
# количество женщин и мужчин
df['gender'].value_counts()
```

```python
# разбивка графика по полу
g = sns.relplot(x='age', y='nbi', data=long_df,hue="gender")
plt.title('Линейная диаграмма с разбивкой по полу', size=20, color='g')
plt.xlabel("возраст")
plt.ylabel("доход от клиента без OPEX");
```
