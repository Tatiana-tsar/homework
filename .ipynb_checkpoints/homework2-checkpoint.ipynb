```python
sber = '3529'
page = 1
num_per_page = 100
moscow = 1 #113
url = f'https://api.hh.ru/vacancies?employer_id={sber}&page={page}&per_page={num_per_page}&area={moscow}'
#url = f'https://api.hh.ru/vacancies?employer_id={sber}&page={page}'
```

```python
res = requests.get(url)
res
```

```python
vacancies = res.json()
num_pages = vacancies.get('pages')
```

```python
# id по всей базе
vac_id=[]
weq=defaultdict(list)
for par in range(0,num_pages):
    url = f'https://api.hh.ru/vacancies?employer_id={sber}&page={page}&per_page={num_per_page}&area={moscow}'
    res = requests.get(url)
    vacancies = res.json()
    vacancy_id = [el.get('id') for el in vacancies.get('items')]
    vac_id.extend(vacancy_id)
    page= par
    
```

```python
# по всем id вытаскиваем из базы информацию в новую таблицу и переводим ее в  DateFrame
kol=0
weq=defaultdict(list)
for id_vac in vac_id:  
    url1=f'https://api.hh.ru/vacancies/{id_vac}?host=hh.ru'
    res1 = requests.get(url1)
    skil = res1.json() 
    weq['id'].append(skil.get('id'))
    weq['name'].append(skil.get('name'))
    weq['skils'].append(skil.get('key_skills'))
    weq['description'].append(skil.get('description'))
    weq['published_at'].append(skil.get('published_at'))
    

```

```python
# из поля вытаскиваем строки
skil=weq.get('skils')
weq['skils']=[','.join([y.get('name') for y in x]) for x in skil]
tabl_new =pd.DataFrame(weq)
tabl_new
```

```python
# Переведите даты публикаций в datetime график опубликованных вакансий по датам
tabl_new.published_at=pd.to_datetime(tabl_new.published_at).dt.date
tabl_new= tabl_new.set_index('published_at')
tabl_new=tabl_new.reset_index()
tabl_new.groupby('published_at')['id'].count().plot(kind='barh');
```

```python
# в дни недели, и определите день недели, в который больше всего публикуют вакансий
tabl_new['weekday']=tabl_new.published_at.apply(lambda x: x.weekday())
```

```python
tabl_new.groupby('weekday')['id'].count()
```

```python
# вакансии с использованием 
name_pr= 'аналитик'  
num_2=tabl_new.name.str.lower().str.contains(name_pr)
tabl1=tabl_new[num_2]
name_pr= 'python'
num_1=tabl1.name.str.lower().str.contains(name_pr)
tablic=tabl1[num_1]
tabl1[num_1].shape
```

```python
# группмровка по вакансиям
tablic.groupby('name')['id'].count()
```

```python
qwe=tablic[tablic.skils.astype(bool)] # удалили пустые строки из столбца
djhs=qwe.skils

skills=djhs.tolist()
kl=Counter(skills)

ku=pd.DataFrame(kl.most_common(),)
```

```python
ku.groupby(0)[1].count().plot(kind='bar', fontsize =5);
plt.xlabel('Количество', color='gray') #barh
plt.ylabel('Вакансии',color='gray')
plt.show()
```

```python
#Задание вытащить все вакансии СБЕРа с ХХ (у апи есть ограничение в 2000, подумайте как его обойти)
obch=[]
for par in range(0,num_pages):
    url = f'https://api.hh.ru/vacancies?employer_id={sber}&page={page}&per_page={num_per_page}&area={moscow}'
    res = requests.get(url)
    vacancies = res.json()
    vacancy_name = [el.get('name') for el in vacancies.get('items')]
    obch.extend(vacancy_name)
    page=par
print(f'Вакансии СБЕРА:',obch)
```
